```python
df_trips = pd.read_csv("/home/ctchervenkov/Documents/data/scenarios/switzerland_2018_10pct/output_trips.csv", sep=";")
df_trips.head()
```

```python
df_trips[
    (df_trips["mode"] == "car")
]["network_distance"].sum() / 780000
```

```python
df_trips_car = df_trips[df_trips["mode"] == "car"]
df_trips_car.head()
```

```python
df_total_distance = df_trips_car[["person_id","network_distance"]].groupby("person_id").sum().reset_index()
df_total_distance.head()
```

```python
print("Average daily car distance:", np.round(df_total_distance["network_distance"].mean()), "km")
```
